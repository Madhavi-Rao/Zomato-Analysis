```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     #for visualization
%matplotlib inline
import seaborn as sns               #for visualization
import warnings
warnings.filterwarnings('ignore')
```

```python
z_df = pd.read_csv('/content/zomato_colby.csv')
z_df
```

```python
z_df.head()
```

```python
#checking what are the variables here:
z_df.columns
```

so now first rename few columns for better understanding of variables -

```python
rename_col = {'country name':'country'}
```

```python
# use a pandas function to rename the current function
z_df = z_df.rename(columns = rename_col)
z_df.head(2)
```

Discarding some of the unwanted columns ---> **Address**, **Locality verbose**,**switch to order menu**,**Longitude**,**Latitude**.

Instead of Address and **Locality Verbose** we can use **Locality** and **City**
,**Switch to order menu** might not provide significant analytical value.

```python
columns_to_drop = ['Address', 'Locality Verbose', 'Switch to order menu','Longitude', 'Latitude',]

# Drop the columns
z_df = z_df.drop(columns=columns_to_drop)
```

```python
z_df.shape
```

```python
#basic information about the dataset
z_df.info()
```

```python
# check duplicate rows in dataset
z_df = z_df.drop_duplicates()
z_df.count()
```

o, there is no any duplicate rows in Dataset

```python
# checking null values of each columns
z_df.isnull().sum()
```

# Data Cleaning and EDA

Only **Cuisine** and **Aggregate Rating** have missing values

```python
# fill missing cuisines with 'Unknown' and drop rows with missing ratings
z_df['Cuisines'].fillna(z_df['Cuisines'].mode()[0], inplace=True)
z_df.dropna(subset=['Aggregate rating'], inplace=True)
```

```python
z_df.info()
```

```python
z_df.head(5)
```

# **Data Visualization**

1) Top 10 cities with most Restaurants

```python
# Top 10 cities with the most restaurants
top_cities = z_df['City'].value_counts().head(10)

plt.figure(figsize=(8, 6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

2) Show distribution of Aggregate Ratings

```python
plt.figure(figsize=(10, 6))
sns.histplot(z_df['Aggregate rating'], bins=30, kde=True)
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

3) What is the Average Cost for Two by Country

```python
avg_cost_country = z_df.groupby('country')['Average Cost for two'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
avg_cost_country.plot(kind='bar')
plt.title('Average Cost for Two by Country')
plt.xlabel('Country')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

4) Number of Reastaurants that offer Online Delivery

```python
online_delivery_count = z_df['Online Delivery'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=online_delivery_count.index, y=online_delivery_count.values)
plt.title('Number of Restaurants Offering Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Count')
plt.show()
```

5) Aggregate Rating by Price Range

```python
avg_rating_price_range = z_df.groupby('Price range')['Aggregate rating'].mean()

plt.figure(figsize=(8, 6))
avg_rating_price_range.plot(kind='bar')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()
```

6) Top 10 Cuisines Offered

```python
top_cuisines = z_df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines Offered')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

7) Distribution of Votes

```python
plt.figure(figsize=(10, 6))
sns.histplot(z_df['Votes'], bins=30, kde=True)
plt.title('Distribution of Votes')
plt.xlabel('Votes')
plt.ylabel('Frequency')
plt.show()
```

8) No. of restaurants with table booking

```python
table_booking_count = z_df['Table Booking'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=table_booking_count.index, y=table_booking_count.values)
plt.title('Number of Restaurants Offering Table Booking')
plt.xlabel('Has Table Booking')
plt.ylabel('Count')
plt.show()
```

9) Average Cost for Two by City

```python
avg_cost_city = z_df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_cost_city.plot(kind='bar')
plt.title('Average Cost for Two by City')
plt.xlabel('City')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

10) Correlation Between Rating and Votes

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Aggregate rating', y='Votes', data=z_df)
plt.title('Correlation Between Rating and Votes')
plt.xlabel('Aggregate Rating')
plt.ylabel('Votes')
plt.show()
```

11) Top 10 Restaurants by Rating

```python
top_rated_restaurants = z_df[['Restaurant Name', 'Aggregate rating']].sort_values(by='Aggregate rating', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_rated_restaurants['Aggregate rating'], y=top_rated_restaurants['Restaurant Name'])
plt.title('Top 10 Restaurants by Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Restaurant Name')
plt.show()
```

12) Distribution of Price Range

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Price range', data=z_df)
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

12) Average Cost for Two by Locality

```python
avg_cost_locality = z_df.groupby('Locality')['Average Cost for two'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_cost_locality.plot(kind='bar')
plt.title('Average Cost for Two by Locality')
plt.xlabel('Locality')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

14) Rating Distribution by City

```python
top_cities = z_df['City'].value_counts().head(10).index
rating_distribution_city = z_df[z_df['City'].isin(top_cities)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='City', y='Aggregate rating', data=rating_distribution_city)
plt.title('Rating Distribution by City')
plt.xlabel('City')
plt.ylabel('Aggregate Rating')
plt.xticks(rotation=45)
plt.show()
```

15) Top 10 Restaurants by Votes

```python
top_voted_restaurants = z_df[['Restaurant Name', 'Votes']].sort_values(by='Votes', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_voted_restaurants['Votes'], y=top_voted_restaurants['Restaurant Name'])
plt.title('Top 10 Restaurants by Votes')
plt.xlabel('Votes')
plt.ylabel('Restaurant Name')
plt.show()
```

16) Average Rating by Cuisine

```python
avg_rating_cuisine = z_df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
avg_rating_cuisine.plot(kind='bar')
plt.title('Average Rating by Cuisine')
plt.xlabel('Cuisines')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```
