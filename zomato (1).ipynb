```python
z_df = pd.read_csv('/content/zomato_colby.csv')
z_df
```

```python
rename_col = {'country name':'country'}
```

```python
# use a pandas function to rename the current function
z_df = z_df.rename(columns = rename_col)
z_df.head(2)
```

Discarding some of the unwanted columns ---> **Address**, **Locality verbose**,**switch to order menu**,**Longitude**,**Latitude**.

Instead of **Address** and **Locality Verbose** we can use **Locality** and **City**
,**Switch to order menu** might not provide significant analytical value.

```python
columns_to_drop = ['Address', 'Locality Verbose', 'Switch to order menu','Longitude', 'Latitude',]

# Drop the columns
z_df = z_df.drop(columns=columns_to_drop)
```

```python
z_df.shape
```

```python
#basic information about the dataset
z_df.info()
```

So, Average cost for two, price range, Aggregate rating and votes are numerical variables

While Restaurant Name, country, city, Locality, Cuisines, Currency, has table booking, has online delivery, is delivering now, rating color, rating text

```python
# check duplicate rows in dataset
z_df = z_df.drop_duplicates()
z_df.count()
```

So, there is no any duplicate rows in Dataset

```python
# checking null values of each columns
z_df.isnull().sum()
```

# Data Cleaning and EDA

Only **Cuisine** and **Aggregate Rating** have missing values

```python
# fill missing cuisines with 'Unknown' and drop rows with missing ratings
z_df['Cuisines'].fillna(z_df['Cuisines'].mode()[0], inplace=True)
z_df.dropna(subset=['Aggregate rating'], inplace=True)
```

```python
z_df.info()
```

```python
z_df.head(5)
```

# **Data Visualization**

1) Top 10 cities with most Restaurants

```python
# Top 10 cities with the most restaurants
top_cities = z_df['City'].value_counts().head(10)

plt.figure(figsize=(8, 6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

2) Show distribution of Aggregate Ratings

```python
plt.figure(figsize=(10, 6))
sns.histplot(z_df['Aggregate rating'], bins=30, kde=True)
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

##### Observation

*   Not rated count is very high
*   Maximum no.of ratings are between 3.9 and 4.2



3) What is the Average Cost for Two by Country

```python
avg_cost_country = z_df.groupby('country')['Average Cost for two'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
avg_cost_country.plot(kind='bar')
plt.title('Average Cost for Two by Country')
plt.xlabel('Country')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

4) Number of Reastaurants that offer Online Delivery

```python
online_delivery_count = z_df['Has Online delivery'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=online_delivery_count.index, y=online_delivery_count.values)
plt.title('Number of Restaurants Offering Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Count')
plt.show()
```

Majority of the restaurants do not offer delivery.

Only 25% of the countries have online delivery.


5) Aggregate Rating by Price Range

```python
avg_rating_price_range = z_df.groupby('Price range')['Aggregate rating'].mean()

plt.figure(figsize=(8, 6))
avg_rating_price_range.plot(kind='bar')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()
```

```python
top_cuisines = z_df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines Offered')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
